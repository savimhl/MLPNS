```python
import numpy as np
import pylab as pl
import scipy
from scipy.spatial.distance import cdist
```

```python
np.random.seed(166)
data = np.random.randn(100,2)
```

```python
label = np.random.randint(0,2, 100)
label
```

```python
ax = pl.figure(figsize =(10,10)).add_subplot(1,1,1)
ax.axis("equal")

ax.plot(data[:,0][label == 0], data[:,1][[label == 0]], "o")
ax.plot(data[:,0][label == 1], data[:,1][[label == 1]], "s")
ax.set_xlabel("x")
ax.set_ylabel("y")
```

```python
testparticle = np.random.randn(2)
```

```python
ax = pl.figure(figsize =(10,10)).add_subplot(1,1,1)
ax.axis("equal")

ax.plot(data[:,0][label == 0], data[:,1][[label == 0]], "o")
ax.plot(data[:,0][label == 1], data[:,1][[label == 1]], "s")
ax.plot(testparticle[0], testparticle[1], "x", ms=30, lw = 3)
ax.set_xlabel("x")
ax.set_ylabel("y")
```

```python
cdist?
```

```python
d = cdist(np.atleast_2d(testparticle), data)   #distanza euclidea
```

```python
np. min(d), d[:,np.argmin(d)]
```

```python
k = 1
labeltest = label[np.argmin(d)]       #assegno colore del più vicino
```

```python
ax = pl.figure(figsize =(10,10)).add_subplot(1,1,1)
ax.axis("equal")


colors=["blue", "orange"]
ax.plot(data[:,0][label == 0], data[:,1][[label == 0]], "o")
ax.plot(data[:,0][label == 1], data[:,1][[label == 1]], "s")
ax.plot(testparticle[0], testparticle[1], "x", ms=30, lw = 3, color = colors[labeltest])
ax.set_xlabel("x")
ax.set_ylabel("y")
```

```python
k = 5
nns = data[np.argsort(d)[0][:k]]
nns
```

```python
nnslabel = label[np.argsort(d)[0][:k]]
nnslabel
```

```python
labeltest2 = 0 if (nnslabel == 0).sum() > (nnslabel == 1).sum() else 1
```

```python
ax = pl.figure(figsize =(10,10)).add_subplot(1,1,1)
ax.axis("equal")


colors=["blue", "orange"]
ax.plot(data[:,0][label == 0], data[:,1][[label == 0]], "o")
ax.plot(data[:,0][label == 1], data[:,1][[label == 1]], "s")
ax.plot(testparticle[0], testparticle[1], "x", ms=30, lw = 3, color = colors[labeltest2])
ax.plot(nns[:,0],nns[:,1], "kx")
ax.set_xlabel("x")
ax.set_ylabel("y")
```

```python
import pandas as pd
```

```python
wines = pd.read_csv("wine.data", header=None)
wines
```

```python
columns = {0:"quality",
           1:"Alcohol",2:"Malic acid",3:"Ash",4:"Alcalinity of ash",5:"Magnesium",
          6:"Total phenols",7:"Flavanoids",8:"Nonflavanoid phenols",9:"Proanthocyanins",
          10:"Color intensity",11:"Hue",12:"OD280/OD315 of diluted wines",13:"Proline"}
wines.rename(columns=columns, inplace=True)
wines
```

```python
from sklearn import preprocessing
```

```python
X = preprocessing.scale(wines.iloc[:,1:], axis = 0)
```

```python
X.mean(axis =0).shape
X.mean(axis=0)
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, wines["quality"], random_state=302)
     

import sklearn.manifold as man #t distributed stochastic neighbourhood embedding
tsne_wines = man.TSNE(n_components=2, perplexity=30.0, early_exaggeration=90.0,
                      random_state=302).fit(X)
wine_proj = tsne_wines.fit_transform(X)
wine_proj_test = tsne_wines.fit_transform(X_test)
     
```

```python

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.inspection import DecisionBoundaryDisplay
_, ax = plt.subplots()


ax.scatter(wine_proj.T[0], wine_proj.T[1], s=10, c = wines["quality"]/3);
ax.set_title("true labels")
     
```

modeling


```python
import sklearn.neighbors as nb
     

nb.KNeighborsClassifier?
     

clf = nb.KNeighborsClassifier(n_neighbors=5)
clf.fit(X_train, y_train)
```

```python
pred = clf.predict(X_test)
pred
```

```python
y_score = clf.predict_proba(X_test)
y_score
     
```

```python



_, ax = plt.subplots()
ax.scatter(wine_proj_test.T[0], wine_proj_test.T[1], s=40, c = y_test/3);
ax.set_title("true labels")

_, ax = plt.subplots(1)

ax.scatter(wine_proj_test.T[0], wine_proj_test.T[1], s=40, c = pred/3);
ax.set_title("predicted labels");
```
