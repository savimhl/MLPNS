```python
from __future__ import print_function
import scipy as sp
import numpy as np
from scipy.optimize import curve_fit, minimize
from scipy import stats
import pylab as pl

%pylab inline
import seaborn
seaborn.set_style("darkgrid")
```

```python
np.random.seed(456)
mymean = 100
```

```python
df = mymean # chi2
p = 0.5    # binomial
n = mymean / p    # mean = n*p binomial

mydist = {}

pl.hist(np.random.chisquare(mymean, size=500))
pl.ylabel("N")
pl.xlabel("x")
```

```python
mysize = (np.random.rand(100) * 2000).astype(int)
mysize[mysize < 10] = 10
pl.plot(mysize,'.')
```

```python
for s in mysize:
  mydist[s] = np.random.chisquare(mymean, size = s)

ax = pl.figure().add_subplot(111)

for d in mydist.keys():
  ax.plot(d, mydist[d].mean(), '.')

ax.plot(pl.xlim(), [100,100], alpha = 0.5)
ax.set_xlabel("Sample size");
ax.set_ylabel("Sample mean");
```
